```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent='Jojo')
```

```python
uniq_add = ['860 United Nation Plaza ', '870 United Nation Plaza ',
       '535 Manhattan Avenue ', '225 East 111th Street ',
       '1 Great Jones Alley ', '1 Fifth Avenue ', '555 Main Street ',
       '575 Main Street ', '1 Avenue B ', '311 East 75th Street ',
       '222 West 14th Street ', '20 Sullivan Street ',
       '516 West 47th Street ', '415 East 52nd Street ',
       '100 Barclay Street ', '2101 Fifth Avenue ',
       '520 East 81st Street ', '385 First Avenue ', '27 Essex Street ']
```

```python
# data['latitude'], data['longitude']=0, 0
```

```python
data.iloc[11995]
```

```python
uni_add = data.loc[11995:,'ADDRESS'].unique().tolist()
```

```python
len(uniq_add)
```

```python
latitude_ = {}
longitude_ = {}
```

```python
failed=set()
```

```python
for i,add in enumerate(uniq_add):
    if i%100 == 0: print('{}/{}'.format(i, len(uniq_add)))
    if (add in latitude_) or (add in failed): continue
    search = add+', Manhattan'
    location = geolocator.geocode(search)
    try:
        latitude_[add]=location.latitude
        longitude_[add]=location.longitude
    except Exception:
        failed.add(add)
```

```python
import pandas as pd
data = pd.read_excel('Sales_SE.xls')
```

```python
for n in data.index:
    add = data.loc[n, 'address']
    if add in latitude_:
        data.loc[n, 'latitude']=latitude_[add]
        data.loc[n, 'longitude']=longitude_[add]
```
